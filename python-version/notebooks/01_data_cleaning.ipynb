```python
import pandas as pd
```

```python
# Load data
employees = pd.read_csv('../../data/employees.csv')
positions = pd.read_csv('../../data/positions.csv')
employments = pd.read_csv('../../data/employments.csv')
candidates = pd.read_csv('../../data/candidates.csv')
recruitments = pd.read_csv('../../data/recruitments.csv')
```

```python
# === Basic Overview ===
print("Candidates:")
display(candidates.head())

print("Employees:")
display(employees.head())

print("Recruitments:")
display(recruitments.head())
```

```python
# === Data Types and Nulls ===
print("Recruitments Info:")
recruitments.info()
print("Missing Values:\n", recruitments.isnull().sum())
```

```python
# === Date Parsing ===
# In Employees
employees['birth_date'] = pd.to_datetime(employees['birth_date'], errors='coerce')
employees['hire_date'] = pd.to_datetime(employees['hire_date'], errors='coerce')
```

```python
# In Employments (rename and parse date)
employments.rename(columns={'effective_date': 'start_date'}, inplace=True)
employments['start_date'] = pd.to_datetime(employments['start_date'], errors='coerce')
```

```python
# === Rename Columns for Consistency ===
candidates.columns = candidates.columns.str.lower()
employees.columns = employees.columns.str.lower()
employments.columns = employments.columns.str.lower()
positions.columns = positions.columns.str.lower()
recruitments.columns = recruitments.columns.str.lower()
```

```python
# === Cleaned Sample Outputs ===
print("\nCleaned Employees:")
display(employees.head(3))

print("\nCleaned Employments:")
display(employments.head(3))
```

```python
# === Save Cleaned Versions ===
employees.to_csv('../outputs/summary_tables/employees_cleaned.csv', index=False)
employments.to_csv('../outputs/summary_tables/employments_cleaned.csv', index=False)
```
